```python
dataset = "test"
table_list_file = f"prepare_data/{dataset}_schemas.csv"
output_file = f"prepare_data/{dataset}_table_list_wo_des.txt"
```

```python
import pandas as pd

df = pd.read_csv(table_list_file)
df
```

```python
table_db_map = df.groupby("db_id")["table"].apply(list).to_dict()
table_db_map
```

```python
table_list_text = ""
for db_id, tables in table_db_map.items():
    table_list_text += f"Database '{db_id}':\n"
    for table in tables:
        table_list_text += f"- {table}\n"
    table_list_text += "\n"

print(table_list_text)
```

```python
with open(output_file, "w+") as f:
    f.write(table_list_text)
```
